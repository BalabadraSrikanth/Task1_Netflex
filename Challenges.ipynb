```python
import pandas as pd
files=pd.read_csv("netflix_titles.csv")

# Retriving the head(0) and assingning to key variable and converting to list
key=files.head(0)
key=list(key)

# Retiving each columns values as a list and adding that list to "aa" list 
value =[]
for i in files:

    a=files[i]
    value.append(list(a))
#print(value)

# usin "Zip()" function converting the key & value lists to dictionry
NetflexDictionary={}
for key,values in zip(key,value):
    for i in files:
        NetflexDictionary[key]=values

#print(NetflexDictionary)
```

```python
'''
## Challenge 1: Total Records

**Task:**
Count and print the total number of records in the dataset.

**Hint:**
Use `len()` on the loaded data list.

print()
'''
for key,value in NetflexDictionary.items():
    print(key,len(value))
```

```python
'''

## Challenge 2: Count Movies vs TV Shows

**Task:**
Count how many entries are Movies and how many are TV Shows.

**Hint:**
Use `row['type']` in a loop. Use `Counter()` or dictionary logic.

'''
a=NetflexDictionary['type']
moviec=0
tvshowc=0
for i in a:
    if i=='Movie':
        moviec+=1
    elif i=='TV Show':
        tvshowc+=1

print("Movie =", moviec, "TV Shows = ", tvshowc)

```

```python
'''
## Challenge 5: Count Titles Per Country

**Task:**
Count how many titles come from each country. If a row lists multiple countries, count each one separately.

**Hint:**
Split `row['country']` using `split(',')`, strip whitespace, and update counts individually.
'''

#print(files["country"])
year=[]
for i in NetflexDictionary['country']:
    year.append(i)
countrys={}
country=[]
for i in year:
    if isinstance(i, str):
        parts = i.split(",")
        for part in parts:
            part=part.strip()
            country.append(part)

country=list(country)
countrys=set(country)
for i in countrys:
    if not i=="":
        print(i,country.count(i))




```
